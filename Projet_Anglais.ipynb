```python
from bs4 import BeautifulSoup 
import pandas as pd
import lxml.html
from  urllib.request import Request, urlopen
import requests
```

```python
'''
Ici je crée Source1 et Source2, car il faut requests.get pour lire en lxml et utiliser l'objet xpath.
Pour source1 il faut la fonction Request pour lire la fonction urlopen(), et par la suite utiliser BeautifulSoup.
'''
```

```python
source1 = Request("https://www.ligue1.fr/classement?seasonId=2014-2015&matchDay=38", headers = {"user-agent" : "Modzilla/5.0"})
```

```python
source2 = requests.get("https://www.ligue1.fr/classement?seasonId=2014-2015&matchDay=38", headers = {"user-agent" : "Modzilla/5.0"})
```

```python
r = urlopen(source1).read()
r
```

```python
soup = BeautifulSoup(r, "html")
soup
```

```python
position = soup.find_all(class_ = "position same")
position1 = soup.find_all(class_ = "position up")
position2 = soup.find_all(class_ = "position down")
print(position, position1, position2)
```

```python
"""
On a 3 classes qui représentent la variable "position".
Voir si la liste à la fin il faudra la mettre en str car l'ensemble de notre de DataFrame sera en String.

"""

position_liste = list()
position_liste1 = list()
position_liste2 = list()

for i in range(len(position)):
    position_liste.append(position[i].text)

for i in range(len(position1)):
    position_liste1.append(position1[i].text)


for i in range(len(position2)):
    position_liste2.append(position2[i].text)

Position_finale = position_liste + position_liste1 + position_liste2
print("La liste non triable :", Position_finale)

# Pour réussir à trier je met la liste en integer.
positon_finale = list(map(int, Position_finale))
positon_finale.sort()
positon_finale
```
